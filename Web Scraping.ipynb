```python
sourse = requests.get('https://en.wikipedia.org/wiki/List_of_state_and_union_territory_capitals_in_India')
sourse = BeautifulSoup(sourse.text,'lxml')
data1 = sourse.find_all('table')
data1
```

```python
data2 = sourse.find('table',class_='wikitable')
print(data2)
```

```python
data2 = data2.find_all('tr')
print(data2)
```

```python
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
G=[]


for i in data2:
    th_data = i.find_all('th')
    td_data = i.find_all('td')
    if len(td_data)==6:
        A.append(td_data[0].text.strip())
        B.append(th_data[0].text.strip())
        C.append(td_data[1].text.strip())
        D.append(td_data[2].text.strip())
        E.append(td_data[3].text.strip())
        F.append(td_data[4].text.strip())
        G.append(td_data[5].text.strip())
        
        
print(A,'\n','\n',B,'\n\n',C,'\n\n',D,'\n\n',E,'\n\n',F,'\n\n',G)    
```

```python
df = pd.DataFrame(zip(B,C,D,E,F,G),columns=["State or UN","Admin Cap","Legis Cap","Judi Cap","Year","Capital"])
                            # columns adds names to dataframe with DataFrame
                            # names adds columns names with read_csv
df
```

```python
df.to_csv('states.csv')
```

### Selenium

```python
from selenium import webdriver
import pandas as pd
# wiki = "https://en.wikipedia.org/wiki/List_of_state_and_union_territory_capitals_in_India"
driver = webdriver.Chrome(executable_path=r'C:\Users\Yogi\Downloads\chromedriver_win32\chromedriver.exe')
driver.get("https://en.wikipedia.org/wiki/List_of_state_and_union_territory_capitals_in_India")
right_table=driver.find_element_by_class_name('wikitable')
# driver.close()
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
G=[]
for row in right_table.find_elements_by_tag_name('tr'):
    cells = row.find_elements_by_tag_name('td')
    states = row.find_elements_by_tag_name('th')
    if len(cells) == 6:
        A.append(states[0].text.strip())
        B.append(cells[1].text.strip())
        C.append(cells[2].text.strip())
        D.append(cells[3].text.strip())
        E.append(cells[4].text.strip())
        F.append(cells[5].text.strip())
        
driver.quit()
dff = pd.DataFrame(zip(B,A,C,D,E,F),columns=["State or UN","Admin Cap","Legis Cap","Judi Cap","Year","Capital"])
dff
```

```python
'''
    find difference between element and elements
    use text to get only text
'''
driver = webdriver.Chrome(executable_path=r'C:\Users\Yogi\Downloads\chromedriver_win32\chromedriver.exe')
driver.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
icc_men_table = driver.find_element_by_class_name('table')
icc_men_table1 = icc_men_table.find_element_by_tag_name('tbody')

rank=[]
team=[]
matches=[]
points=[]
rating=[]

for row in icc_men_table1.find_elements_by_tag_name('tr'):
    icc_data = row.find_elements_by_tag_name('td')
    rank.append(icc_data[0].text.strip())
    team.append(icc_data[1].text.strip())
    matches.append(icc_data[2].text.strip())
    points.append(icc_data[3].text.strip())
    rating.append(icc_data[4].text.strip())
driver.quit()
icc_df = pd.DataFrame(zip(rank,team,matches,points,rating),columns=['Rank','Team','Matches','Points','Rating'])
icc_df
```

## scraping with paging

### by adjusting url

```python
address=[]
item=[]
item_required=[]
start_date=[]
end_date=[]

driver = webdriver.Chrome(executable_path=r'C:\Users\Yogi\Downloads\chromedriver_win32\chromedriver.exe')
for i in range(1,5):
    if i==1:
        driver.get("https://bidplus.gem.gov.in/bidlists")
        bid_table = driver.find_element_by_id('pagi_content')
        bid_table =  bid_table.find_elements_by_class_name('border')
        for i in bid_table:
            item.append(i.find_elements_by_tag_name('span')[0].text.strip())
            item_required.append(i.find_elements_by_tag_name('span')[1].text.strip())
            start_date.append(i.find_elements_by_tag_name('span')[2].text.strip())
            end_date.append(i.find_elements_by_tag_name('span')[3].text.strip())
            address.append(i.find_element_by_class_name('add-height').text.strip())
           
    else:
        bid="https://bidplus.gem.gov.in/bidlists?bidlists&page_no="+str(i)
        driver.get(bid)
        bid_table = driver.find_element_by_id('pagi_content')
        bid_table = bid_table.find_elements_by_class_name('border')
        for i in bid_table:            
            item.append(i.find_elements_by_tag_name('span')[0].text.strip())
            item_required.append(i.find_elements_by_tag_name('span')[1].text.strip())
            start_date.append(i.find_elements_by_tag_name('span')[2].text.strip())
            end_date.append(i.find_elements_by_tag_name('span')[3].text.strip())
            address.append(i.find_element_by_class_name('add-height').text.strip())
driver.quit()
            

bid_df = pd.DataFrame(zip(item,item_required,address,start_date,end_date),columns=['item','item_required','address','start_date','end_date'])
```

### by click to  next page

```python
from selenium import webdriver
import pandas as pd
address=[]
item=[]
item_required=[]
start_date=[]
end_date=[]

#driver = webdriver.Chrome(executable_path=r'C:\Users\Yogi\Downloads\chromedriver_win32\chromedriver.exe')
driver = webdriver.Chrome(executable_path=r'C:\Users\Yogi\Downloads\chromedriver_win32\chromedriver.exe')
driver.get("https://bidplus.gem.gov.in/bidlists")
c=0
while(c<=20): 
    c+=1
    bid_table = driver.find_element_by_id('pagi_content')
    bid_table = bid_table.find_elements_by_class_name('border')
    for i in bid_table:             
            item.append(i.find_elements_by_tag_name('span')[0].text.strip())
            item_required.append(i.find_elements_by_tag_name('span')[1].text.strip())
            start_date.append(i.find_elements_by_tag_name('span')[2].text.strip())
            end_date.append(i.find_elements_by_tag_name('span')[3].text.strip())
            address.append(i.find_element_by_class_name('add-height').text.strip())
    driver.find_element_by_xpath("//a[@rel='next']").click()
driver.quit() 
dff = pd.DataFrame(zip(item,item_required,address,start_date,end_date),columns=['Item','Item Reuired','Address','Start Date','End Date'])
dff
```
